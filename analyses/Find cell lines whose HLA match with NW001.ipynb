```r
library('tidyverse')
```

```r
depmap_tbl <- 
    read_csv('../potency/cell_line_selection/Model_23Q2.csv')

depmap_tbl %>%
    head(10)
```

```r
hla_tbl <-
    read_tsv(
        '../hla/allelefreq_rare_list/tron_cell_line_portal.tsv',
        col_names=c('StrippedCellLineName', 'allele_type', 'allele_1', 'allele_2')
    ) %>%
    mutate(
        `allele_1`=paste(`allele_type`, `allele_1`, sep='*'),
        `allele_2`=paste(`allele_type`, `allele_2`, sep='*')
    ) %>%
    pivot_longer(
        names_to='allele_no',
        values_to='allele_4digit',
        c(`allele_1`, `allele_2`)
    ) %>%
    select(-`allele_no`) %>%
    distinct() %>%
    filter(`allele_type` %in% c('A', 'B', 'C'))

hla_tbl %>%
    head()
```

```r
a1101_lines <-
    left_join(depmap_tbl, hla_tbl, by='StrippedCellLineName') %>%
    filter(`allele_4digit` == 'A*11:01')

a1101_lines
```

```r
a1101_lines %>%
    write_csv('a1101_lines.csv', na='')
```
