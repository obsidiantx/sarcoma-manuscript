```r
library('tidyverse')

org_opts <- options()
```

```r
mait_cdr3_scores <-
    read_tsv('MAITMATCH_results_20240502.tsv') %>%
    filter(`Score` > 0.90)

mait_cdr3_scores
```

```r
rep_tra <- 
    read_csv('../../tcr/nw001/iRepertoire-170311_2306-RD-0017_Release_/atlas/IL15_REP_TIL__TRA.csv') %>%
    rename(`CDR3`=`junction_aa`) %>%
    group_by(`CDR3`) %>%
    summarize(`duplicate_count`=sum(`duplicate_count`)) %>%
    ungroup() %>%
    mutate(
        `read_fraction`=`duplicate_count`/sum(`duplicate_count`),
        `fraction_str`=scales::percent(`read_fraction`, accuracy=.1)
    ) %>%
    select(`CDR3`, `fraction_str`, `read_fraction`) %>%
    arrange(-`read_fraction`) %>%
    mutate(`is_MAIT`=`CDR3` %in% mait_cdr3_scores$`Sequence`)

rep_tra
```

```r
rep_tra %>%
    group_by(`is_MAIT`) %>%
    summarize(
        `no_of_clonotypes`=n(),
        `total_fraction`=sum(`read_fraction`),
        `total_fraction_str`=scales::percent(`total_fraction`, accuracy=.1),
        .groups='drop'
    ) %>%
    mutate(
        `fraction_of_clonotyes`=`no_of_clonotypes`/sum(`no_of_clonotypes`),
        `fraction_of_clonotyes_str`=scales::percent(`fraction_of_clonotyes`, accuracy=.1)
    )

    
```
